```python
import pandas as pd

movies_df = pd.read_csv('data\movies.csv')
movies_df
```

```python
mov_df = movies_df.copy()
mov_df['movieId'].nunique()
```

```python
rat1_df = pd.read_csv(r'data\ratings1.csv')
rat1_df['userId'].nunique()
```

```python
rat2_df = pd.read_csv(r'data\ratings2.csv')
rat2_df['userId'].nunique()
```

```python
date_df = pd.read_csv(r'data\dates.csv')
date_df
```

```python
date_df['date']=pd.to_datetime(date_df['date'])
date_df['date'].dt.year.value_counts().index[0]
```

```python
ratings = pd.concat([rat1_df, rat2_df])
display(ratings)
```

```python
ratings = pd.concat(
    [rat1_df, rat2_df],
    ignore_index=True
)
display(ratings)
```

```python
print('Число строк в таблице ratings: ', ratings.shape[0])
print('Число строк в таблице dates: ', date_df.shape[0])
print(ratings.shape[0] == date_df.shape[0])
```

```python
display(rat1_df.tail(1))
display(rat2_df.head(1))
```

```python
ratings = ratings.drop_duplicates(ignore_index=True)
ratings
```

```python
ratings_dates = pd.concat([ratings, date_df], axis=1)
display(ratings_dates.tail(1))
```

```python
mov_df
```

```python
def full_movie_list(mov_id):
    for index,movie_id in enumerate(movies_df['movieId']):
        if mov_id == movie_id:
            return movies_df.iloc[index, 1:]
    
movies_df_full = ratings['movieId'].apply(full_movie_list)
type(movies_df_full)
movies_df_full.info()
```

```python
type(movies_df_full)
movies_df_full.info()
```

```python
ratings_mov = pd.concat([ratings, movies_df_full], axis=1)
display(ratings_mov)
```

```python
joined_false = ratings_dates.join(
    mov_df,
    rsuffix='_right',
    how='inner'
)
display(joined_false)
```

```python
joined = ratings_dates.join(
    mov_df.set_index('movieId'),
    on='movieId',
    how='left'
)
display(joined)
```

```python
merged = ratings_dates.merge(
    mov_df,
    on='movieId',
    how='left'
)
display(merged.head())
```
